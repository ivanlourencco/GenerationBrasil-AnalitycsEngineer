```python
import requests
import json
import pandas as pd

url = 'https://api.open-meteo.com/v1/forecast'


# perguntando ao usuario os parametros que iremos usar
lat = float(input("Digite a latitude da cidade (ex: -23.5505 para SP): "))
lon = float(input("Digite a longitude da cidade (ex: -46.6333 para SP): "))
quantos_dias = int(input("Quantos dias passados você quer consultar? (ex: 7): "))

parametros = {
    'latitude' : lat,
    'longitude' : lon,
    'daily' : ['temperature_2m_max','precipitation_sum'],
    'timezone' : 'auto',
    'past_days' : quantos_dias
}

# Conectando a API e pedindo pelos dados
resposta = requests.get(url, params=parametros)

if resposta.status_code == 200:
    print("Dados recebidos com sucesso!")

    # Convertemos a resposta para um dicionário Python (formato JSON)
    dados = resposta.json()
    dados_diarios = dados['daily']

    #criar o dataframe com o pandas
    df_dados_clima = pd.DataFrame(dados_diarios)



  #transformando em dados dataframe
    df_dados_clima = pd.DataFrame(dados_diarios)

  # EXTRA - mostrar no padrao dos EUA
  # (temperatura * 1.8) + 32
    df_dados_clima['temperature_2m_max_fahrenheit'] = ((df_dados_clima['temperature_2m_max'] * 1.8) + 32).round(2)


    display(df_dados_clima)

  #transformando em csv
    df_dados_clima.to_csv('relatorio_prof_luiz.csv', index=False)

else:
  print(f"Erro na requisição: {resposta.status_code}")

```
